```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
movies_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projects/Movie_Recommendation_System/notebooks/preprocessing/preprocessed_dataset/preprocessed_text_dataset.csv')
```

```python
id_vector_map = {}
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/Projects/Movie_Recommendation_System/config/id_vector_map.pkl','rb') as file:
  id_vector_map = pickle.load(file)
```

```python
def print_recommendation(movie_name,n):
  try:
    movie_name = movie_name.strip().lower()
    temp_df = movies_df[movies_df['title']==movie_name]
    # movie_vector = id_vector_map[temp_df['id'].to_list()[0]].reshape(1,-1)
    movie_vector = id_vector_map[temp_df['id'].to_list()[0]]

    temp_cosines = []

    for movie_id,vector in id_vector_map.items():
      # temp_cosines.append((movie_id,1 - cosine(movie_vector, vector)))
      temp_cosines.append((movie_id,cosine_similarity(movie_vector,vector)))
    temp_cosines.sort(key=lambda x: x[1],reverse=True)
    print(f'Movies similar to "{movie_name}"')

    count=0
    for ind,elem in enumerate(temp_cosines):
      if ind == 0:
        continue
      if count==n:
        break
      count+=1

      row = movies_df[movies_df['id']==elem[0]]
      movie_names = row['title'].to_list()[0]
      cosine_value = round(elem[1][0][0]*100,2)

      print(f'{ind}. {movie_names.title()} (similarity: {cosine_value}%)')
  except IndexError:
    print('Unfortunately this movie is not present in our dataset. Try again with different movie')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/id_word_vector_map.pkl','rb') as file:
  id_word_vector_map = pickle.load(file)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
try:
  og_mov_name = 'toy story'
  mov_key = ''.join(og_mov_name.strip().split())
  print(f'Movies similar to: "{og_mov_name}"')
  mov_vector = id_word_vector_map[title_id_mapping[mov_key]].reshape(1,-1)
  if mov_vector is None:
    raise KeyError
  t_id_vect_map = {}
  for id,vector in id_word_vector_map.items():
    t_id_vect_map[id] = cosine_similarity(mov_vector,vector.reshape(1,-1))
  movies = []
  counter = 0
  for id,cosine_value in sorted(t_id_vect_map.items(),key=lambda x: x[1],reverse=True):
    if counter == 6:
      break
    counter+=1
    movies.append(id_title_mapping[id])
  for ind,movie_name in enumerate(movies):
    if ind==0:
      continue
    print(f'{ind}. {movie_name}')
except KeyError:
  print('Sorry this movie does not exists in our database. Try again with different movie.')
```

## Feature Engineering

```python
def get_genres(row):
    row = ast.literal_eval(row)
    genres = []
    for dictionary in row:
        genres.append(dictionary['name'].lower())
    return ' '.join(genres) if len(genres)>0 else None

def get_keyword(row):
    row = ast.literal_eval(row)
    keywords = []
    for dictionary in row:
        keywords.append(dictionary['name'].lower())
    return ' '.join(keywords) if len(keywords)>0 else None

def get_crew(row):
    director = []
    row = ast.literal_eval(row)
    for dictionary in row:
        if dictionary['job']=='Director':
            director_name = dictionary['name'].lower().split()
            director.append(''.join(director_name))
            break
    return ' '.join(director) if len(director)>0 else None

def get_cast(row):
    row = ast.literal_eval(row)
    actors = []
    for dictionary in row[:5]:
        actor_name = dictionary['name'].lower().split()
        actors.append(''.join(actor_name))
    return ' '.join(actors) if len(actors)>0 else None
```
